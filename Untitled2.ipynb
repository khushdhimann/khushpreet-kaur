```python
#importing datasets
import pandas as pd
data = pd.read_csv('survey lung cancer (1).csv')
data.head(5)
data.drop('GENDER', axis=1, inplace=True)
data.head()
```

```python
data.LUNG_CANCER = [1 if each=="YES" else 0 for each in data.LUNG_CANCER]
```

```python
# refining the data
data.duplicated().sum()
# removing duplicates
data.drop_duplicates(inplace=True)
#duplicates removed
data.duplicated().sum()
#check for null values
data.isna().sum()
```

```python
# creating the feature and the output vectors
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
import sklearn.model_selection

# dividing the data into training and test sets
tts = sklearn.model_selection.train_test_split
x_train , x_test , y_train , y_test = tts(x,y,test_size = 0.6, random_state = 6)
```

```python
# implementing the classifier
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
# train the model
gnb.fit(x_train,y_train)
# make prediction
gnb_prediction = gnb.predict(x_test)
# accuracy of the model
print("the accuracy of the model is ",sklearn.metrics.accuracy_score(y_test,gnb_prediction))
```

```python
# using decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier(random_state = 0)
# train the model
dc.fit(x_train,y_train)
dc_pred = dc.predict(x_test)
# accuracy
print("accuracy of the model is ",sklearn.metrics.accuracy_score(y_test,dc_pred))
```

```python
#using svm
from sklearn.svm import SVC
sv = SVC(kernel= "linear")
sv.fit(x_train,y_train)
sc_pred  = sv.predict(x_test)
# print acuuracy
print("the accuracy is :",sklearn.metrics.accuracy_score(y_test,sc_pred))
```

```python
from sklearn.linear_model import LogisticRegression
# using logistic regression
lr = LogisticRegression(max_iter =5000)
lr.fit(x_train,y_train)
lr_pred = lr.predict(x_test)
# accuracy_score
print("the accuracy is :",sklearn.metrics.accuracy_score(y_test,lr_pred))
```
